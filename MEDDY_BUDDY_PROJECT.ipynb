## 1. Does gender matter for policy claims?


    new_code_cell
# Statistical test for gender difference
male_charges = df[df['sex'] == 'male']['charges in INR']
female_charges = df[df['sex'] == 'female']['charges in INR']
t_stat, p_value = stats.ttest_ind(male_charges, female_charges)

print(f"Mean charges in INR by gender:")
print(df.groupby("sex")["charges in INR"].mean())
print(f"\nT-test p-value: {p_value:.4f}")
print(f"Statistically significant difference: {p_value < 0.05}")

ðŸ” Key Insights from EDA

1 Gender has minimal impact on insurance claims and should not be a constraint for policy issuance.

2 The average insurance cost per policy provides a reliable baseline for pricing strategies.

3 Geographic location shows moderate variation in claims, suggesting region-based policy customization may be beneficial.

4 The number of dependents slightly increases claim amounts but is not a dominant factor.

5 BMI has a strong positive correlation with insurance charges, indicating higher risk for unhealthy individuals.

6 Smoking status is the most influential factor, with smokers claiming significantly higher amounts than non-smokers.

7 Age shows a steady increase in claim value, but it is not a strict barrier to insurance coverage.

8 BMI-based discounts can be introduced to encourage healthier lifestyles and reduce long-term claim risks.
